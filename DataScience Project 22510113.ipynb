```python
# Normalization
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = df[[' income_annum',' loan_amount', ' cibil_score']]
x_scaled = scaler.fit_transform(x)
x_scaled.ndim
type(x_scaled)
```

```python
### Outlier detection and removal
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination = 0.05, random_state = 0)
```

```python
x_encoded = x.apply(le.fit_transform)
clean = iso.fit_predict(x_encoded,y)
```

```python
x_clean = x[clean==1]
y_clean = y[clean==1]
```

```python
x_clean.shape, y_clean.shape
```

```python
## Feature selection
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(k=3, score_func = chi2)
```

```python
skf.fit_transform(x_clean,y_clean)
```

```python
skf.get_support()
```

```python
x_clean = x_clean.iloc[:, skf.get_support()]
```

```python
x_clean = x_clean.iloc[:, skf.get_support()]
```

```python
x_clean
```

```python
### Feature Scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x_clean)
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
## Cross-Validation
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test = train_test_split(x_scaled,y_clean,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
## Build the model
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state = 0)
```

```python
clf.fit(x_train, y_train)
```

```python
### Performance Analysis
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
